### 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
# visualize the dataframe using seaborn-pairplot to check for any relationship among SL,SW,PL and PW.
sns.pairplot(df, hue="Classification")
plt.show()
```

#### All the variables are forming a cluster in case of Iris-versicolor(orange) and Iris-virginica(green). However, values of iris-setosa stands apart. There is a linear relationship between PW and PL.

```python
# distribution of SL,SW,PL,PW using histplot

fig, axes = plt.subplots(2,2)

sns.histplot(df["SL"], color="green", ax=axes[0,0])
sns.histplot(df["SW"], ax=axes[0,1])
sns.histplot(df["PL"], color="red", ax=axes[1,0])
sns.histplot(df["PW"], color="orange", ax=axes[1,1])
plt.tight_layout()
plt.show()
```

#### Most of the irises have: Sepal Length of 3-6 units, Sepal width around 3 units , Petal length around 1 unit  and Petal width below 0.5 units.

```python
# box plot to show how sepal width varies for different types of irises
sns.boxplot(y=df["SW"],x=df["Classification"],  hue=df["Classification"])

plt.legend()
plt.show()
```

#### The median sepal width is highest for iris-setosa 

```python
# check for relationship between Sepal length and Petal Length using lmplot for all three classifications
sns.lmplot(x="SL",y="PL", hue="Classification", data=df)
plt.title("SL vs PL")
plt.show()
```

#### There is a positive linear relationship between SL vs PL for classes Iris-versicolor and Iris-virginica

```python
# check for relationship between Sepal length and Sepal Width using lmplot
sns.lmplot(x="SL",y="SW", hue="Classification", data=df)
plt.title("SL vs SW")
plt.show()
```

#### SL vs SW  regression plot shows positive linear relationship for all three irises.

```python
# check for relationship between Petal Width and Petal Length using lmplot
sns.lmplot(x="PW",y="PL", hue="Classification", data=df)
plt.title("PW vs PL")
plt.show()
```

#### PL vs PW shows that there is evident positive relationship in iris-versicolor and iris-virginica.
